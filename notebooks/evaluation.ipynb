2)	Load a lightweight GPT model (distilgpt2)

```
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load tokenizer and model
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Add padding token (GPT2 models have none)
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.eos_token_id
```

3)	Build the chatbot function

```
def chatbot(prompt):
    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        padding=True,
        truncation=True
    )

    output = model.generate(
        **inputs,
        max_length=150,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.8,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```

```
# evaluation_scripts.py functionality inside Colab

import torch
import math
import numpy as np
from transformers import AutoTokenizer, AutoModelForCausalLM
from sacrebleu import corpus_bleu
from tqdm import tqdm
import csv

# reproducibility
torch.manual_seed(42)
np.random.seed(42)

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# load prompts and references
with open("validation_prompts.txt", "r") as f:
    prompts = [line.strip() for line in f if line.strip()]

with open("references.txt", "r") as f:
    references = [line.strip() for line in f if line.strip()]

responses = []
nlls = []

for p in tqdm(prompts, desc="Generating responses"):
    enc = tokenizer(p, return_tensors="pt").to(device)
    with torch.no_grad():
        out = model.generate(
            **enc,
            max_length=150,
            do_sample=True,
            top_k=50,
            top_p=0.95,
            temperature=0.8,
            pad_token_id=tokenizer.eos_token_id
        )

    resp = tokenizer.decode(out[0], skip_special_tokens=True)
    responses.append(resp)

    # compute response-only NLL
    ids_prompt = tokenizer(p, return_tensors="pt").input_ids.to(device)
    ids_resp = tokenizer(resp, return_tensors="pt").input_ids.to(device)
    full_ids = torch.cat([ids_prompt, ids_resp], dim=1)

    labels = full_ids.clone()
    labels[0, :ids_prompt.shape[1]] = -100

    with torch.no_grad():
        out_loss = model(full_ids, labels=labels)
        nlls.append(out_loss.loss.item())

# Save CSV
with open("results_responses.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["prompt", "response", "nll"])
    for p, r, n in zip(prompts, responses, nlls):
        writer.writerow([p, r, n])

print("Created: results_responses.csv")

# Metrics
avg_nll = sum(nlls) / len(nlls)
perplexity = math.exp(avg_nll)
bleu = corpus_bleu(responses, [references]).score

print("Average NLL:", avg_nll)
print("Perplexity:", perplexity)
print("BLEU:", bleu)

# Save metrics text file
with open("evaluation_metrics.txt", "w") as f:
    f.write(f"avg_nll,{avg_nll}\nperplexity,{perplexity}\nbleu,{bleu}\n")

print("Created: evaluation_metrics.txt")
```
